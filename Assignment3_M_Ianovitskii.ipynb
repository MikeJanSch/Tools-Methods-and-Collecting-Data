```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
janis= requests.get("https://www.azlyrics.com/lyrics/janisjoplin/blindman.html")
```

```python
janis = requests.get("https://www.azlyrics.com/lyrics/janisjoplin/blindman.html")
joplin= janis.text
print(joplin)
```

```python
document = BeautifulSoup(joplin, "html.parser")
print(document.text)
```

```python
document.find('h1').text
```

```python
title=document.find('h1').text[1:9]
title
```

```python
lyrics= document.find("div",attrs={'class':"col-xs-12 col-lg-8 text-center"}).text.strip().replace("\r", "").split('\n')
while '' in lyrics:
    lyrics.remove('')
lyrics=lyrics[4:27]
lyrics
```

```python
lyrics="".join(lyrics)
lyrics
```
